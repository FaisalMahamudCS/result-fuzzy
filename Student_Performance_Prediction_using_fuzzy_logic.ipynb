```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
marks = ctrl.Antecedent(np.arange(0,76,1), 'marks')
attendence=ctrl.Antecedent(np.arange(0,11,1), 'attendence')
assignment=ctrl.Antecedent(np.arange(0,16,1), 'assignment')
performance=ctrl.Consequent(np.arange(0,105,5), 'performance')
```

```python
marks['poor']=fuzz.trimf(marks.universe, [0, 0, 30])
marks['medium']=fuzz.trimf(marks.universe, [0, 30, 75])
marks['good']=fuzz.trimf(marks.universe, [55, 75, 75])
```

```python
marks.view()
```

```python
attendence['poor']=fuzz.trimf(attendence.universe, [0,0, 4 ])
attendence['medium']=fuzz.trimf(attendence.universe, [0, 5,10])
attendence['good']=fuzz.trimf(attendence.universe, [7, 10, 10])
```

```python
attendence.view()
```

```python
assignment['poor']=fuzz.trimf(assignment.universe, [0,0,7])
assignment['medium']=fuzz.trimf(assignment.universe, [0,7,15])
assignment['good']=fuzz.trimf(assignment.universe, [8, 15, 15])
```

```python
assignment.view()
```

```python
performance['poor']=fuzz.trimf(performance.universe, [0,0,45])
performance['medium']=fuzz.trimf(performance.universe, [0,45,100])
performance['good']=fuzz.trimf(performance.universe, [70, 100, 100])
```

```python
performance.view()
```

```python
rule1 = ctrl.Rule(marks['poor'] & attendence['poor'] & assignment['poor'], performance['poor'])
rule2 = ctrl.Rule(marks['poor'] & attendence['poor'] & assignment['medium'], performance['poor'])
rule3 = ctrl.Rule(marks['poor'] & attendence['poor'] & assignment['good'], performance['poor'])
rule4 = ctrl.Rule(marks['poor'] & attendence['medium'] & assignment['poor'], performance['poor'])
rule5 = ctrl.Rule(marks['poor'] & attendence['medium'] & assignment['medium'], performance['poor'])
rule6 = ctrl.Rule(marks['poor'] & attendence['medium'] & assignment['good'], performance['poor'])
rule7 = ctrl.Rule(marks['poor'] & attendence['good'] & assignment['poor'], performance['poor'])
rule8 = ctrl.Rule(marks['poor'] & attendence['good'] & assignment['medium'], performance['poor'])
rule9 = ctrl.Rule(marks['poor'] & attendence['good'] & assignment['good'], performance['poor'])
rule10 = ctrl.Rule(marks['medium'] & attendence['poor'] & assignment['poor'], performance['medium'])
rule11= ctrl.Rule(marks['medium'] & attendence['poor'] & assignment['medium'], performance['medium'])
rule12 = ctrl.Rule(marks['medium'] & attendence['poor'] & assignment['good'], performance['medium'])
rule13 = ctrl.Rule(marks['medium'] & attendence['medium'] & assignment['poor'], performance['medium'])
rule14= ctrl.Rule(marks['medium'] & attendence['medium'] & assignment['medium'], performance['medium'])
rule15 = ctrl.Rule(marks['medium'] & attendence['medium'] & assignment['good'], performance['medium'])
rule16 = ctrl.Rule(marks['medium'] & attendence['good'] & assignment['poor'], performance['medium'])
rule17= ctrl.Rule(marks['medium'] & attendence['good'] & assignment['medium'], performance['medium'])
rule18= ctrl.Rule(marks['medium'] & attendence['good'] & assignment['good'], performance['medium'])
rule19= ctrl.Rule(marks['good'] & attendence['poor'] & assignment['poor'], performance['medium'])
rule20= ctrl.Rule(marks['good'] & attendence['poor'] & assignment['medium'], performance['medium'])
rule21= ctrl.Rule(marks['good'] & attendence['poor'] & assignment['good'], performance['good'])
rule22 = ctrl.Rule(marks['good'] & attendence['medium'] & assignment['poor'], performance['good'])
rule23 = ctrl.Rule(marks['good'] & attendence['medium'] & assignment['medium'], performance['good'])
rule24 = ctrl.Rule(marks['good'] & attendence['medium'] & assignment['good'], performance['good'])
rule25 = ctrl.Rule(marks['good'] & attendence['good'] & assignment['poor'], performance['good'])
rule26= ctrl.Rule(marks['good'] & attendence['good'] & assignment['medium'] , performance['good'])
rule27 = ctrl.Rule(marks['good'] & attendence['good'] & assignment['good'], performance['good'])
```

```python
performance_ctrl = ctrl.ControlSystem([rule1, rule2, rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26,rule27])
```

```python
performance1=ctrl.ControlSystemSimulation(performance_ctrl)

performance1.input['marks']=50
performance1.input['attendence']=6
performance1.input['assignment']=13
performance1.compute()

```

```python
print(performance1.output['performance'])
performance.view(sim=performance1)
```
